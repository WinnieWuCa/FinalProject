# General Restaurant Info

### Retrieve General Info

```python
# Identify and save the link to the restaurant's details
counter = 0
biz_names = []
biz_addrs=[]
yelp_links = []
ratings = []
food_types = []

start_page = 0
#There is a total of 1162 gluten free restaruants returend in Yelp search results. We will retrieve reviews of the top 
# 300 restaurants (based on review counts) to see if they truly offer gluten free choices 
while start_page<=301:
    url = "https://www.yelp.ca/search?find_desc=Gluten-Free&find_loc=Toronto%2C%20ON&ns=1&sortby=review_count&start="+str(start_page) 
    #print(url) 
    browser.visit(url)

    # Retrieve page with the requests module
    html = browser.html
    # Create BeautifulSoup object; parse with 'html.parser'
    soup = bs(html, 'html.parser')
    # Retrieve the parent divs for all the reviews

    results = soup.find_all(class_='lemon--div__373c0__1mboc mainAttributes__373c0__1r0QA arrange-unit__373c0__1piwO arrange-unit-fill__373c0__17z0h border-color--default__373c0__2oFDT')

    for result in results:
        #print(result.prettify())
        counter = counter + 1
        
        #Business Name
        details= result.find(class_="lemon--a__373c0__IEZFH link__373c0__29943 link-color--blue-dark__373c0__1mhJo link-size--inherit__373c0__2JXk5")
        name = details["name"]   
        biz_names.append(name) 
        
        #yelp Internal Site
        yelp_link = details['href']
        yelp_links.append(yelp_link)
        
        #Overall Rating
        details = result.find_all(class_="lemon--div__373c0__1mboc")
        if details is None:
            print("no rating 1")
            rating=""
            ratings.append(rating)
        else:
            for item in details:
                try:
                    rating = item["aria-label"]
                    ratings.append(rating)
                except:
                    print("no rating 1")
                    print(name)
                    rating=""
                    ratings.append(rating)
                    #continue
                
        #Business Address     
        details = result.find('span',{'class':'lemon--span__373c0__3997G'})
        biz_addr = details.text
        #print("#Business Address ")
        #print(biz_addr)
        biz_addrs.append(biz_addr) 
        
        #Type of Food
        details = result.find('a', {'class':"lemon--a__373c0__IEZFH link__373c0__29943 link-color--inherit__373c0__15ymx link-size--default__373c0__1skgq"}) 
        food_type = details.text
        food_types.append(food_type) 
       
    start_page =start_page+30
    #---------------------------------------------------#
        #Close browser
        #exit()
        
    
print("counter")
print(counter)

```

# Restaurant Details

```python
biz_counter = 0

biz_revieweds =[]
biz_tels=[]
biz_Websites=[]
biz_lats = []
biz_longs = []
biz_reviews=[]
biz_hoods=[]

for address in yelp_links: 
    biz_reviewed = biz_names[biz_counter]
     
    executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
    browser = Browser('chrome', **executable_path, headless=False)

    #internal url
    url = address
    #print("https://www.yelp.com" + url)
    browser.visit("https://www.yelp.com" + url)
    # Retrieve page with the requests module
    html = browser.html
    # Create BeautifulSoup object; parse with 'html.parser'
    soup = bs(html, 'html.parser')
    #print(soup) 
    #Business Tel
    #details =soup.find('p',{'class':'lemon--p__373c0__3Qnnj text__373c0__2pB8f text-color--normal__373c0__K_MKN text-align--left__373c0__2pnx_'})
    details =soup.find('span',{'class':'biz-phone'})
    if details is None:
        print("no tel")
        biz_tel=""
    else:
        biz_tel= details.text
    biz_tels.append(biz_tel) 
    
    #Business Website
    #details = soup.find('span', {'class':"biz-website js-biz-website js-add-url-tagging"})
    #biz_Website = details.find('a')['href'].text
    #print(details.find('a')['href'])
    #print("biz_Website")
    #print(biz_Website)
    #biz_Website = details.text
    #biz_Websites.append(biz_Website) 
    
    #print(biz_Website)
    
    #Business location latitude and longitude
    details = soup.find('div', class_="lightbox-map hidden")
    if details is None:
        print("no latitude and longitude")
        biz_lat=""
        biz_long="" 
    else:
        #print(details["data-map-state"])
        j = json.loads(details["data-map-state"])
        biz_location=j["markers"][1]['location']
        biz_lat=biz_location["latitude"]
        biz_long=biz_location["longitude"]
    biz_lats.append(biz_lat) 
    biz_longs.append(biz_long) 
    
    #-----------------------Hoods-----------------------#
    #neighborhood-str-list
    details =soup.find('span',{'class':'neighborhood-str-list'})
    if details is None:
        print("no hood info")
        biz_hood=""
    else:
        biz_hood = details.text
    biz_hoods.append(biz_hood)
    
    #-----------------------Review of the restaurant-----------------------#
    # Retrieve the parent divs for all the reviews
    results = soup.find_all("div", class_='review-content') 
    biz_review_counter=0
    #print(results)
    for result in results:  
        biz_review = result.p.text
        biz_reviews.append(biz_review)
        biz_review_counter=biz_review_counter+1
        
        biz_revieweds.append(biz_reviewed)
        
    biz_counter =biz_counter+1
    
    #Close browser
    #exit()
print('========done==========') 
print("biz_tels")
print(biz_tels)
print("biz_lats")
print(biz_lats)
print("biz_longs")
print(biz_longs)
print("biz_hood")
print(biz_hoods)
print('========done==========') 
```
